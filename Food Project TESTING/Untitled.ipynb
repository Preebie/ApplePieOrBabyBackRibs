```python
file_names = ["applepie" + str(x) + ".jpg" for x in range(1, 9)]
print(file_names)
images = []
for x in file_names:
    img = cv2.imread("images/" + x)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img)
```

```python
images = np.array(images)
```

```python
plt.imshow(images[5])
```

```python
def scale_resize_image(image):
    image = tf.image.convert_image_dtype(image, tf.float32) # equivalent to dividing image pixels by 255
    image = tf.image.resize(image, (128, 128)) # Resizing the image to 224x224 dimention
    return image
resized_images = []
for x in images:
    x=scale_resize_image(x)
    resized_images.append(x)
```

```python
plt.imshow(resized_images[5])
```

```python
resized_images = np.array(resized_images)
```

```python
loaded_model= tf.keras.models.load_model('./model.h5')
```

```python
pred = loaded_model.predict(resized_images)
for x in pred:
    if x > .5:
        print("babybackribs")
    else:
        print("applepie")
```
